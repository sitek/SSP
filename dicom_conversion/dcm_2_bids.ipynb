```python
import os
import shutil
import re
import json

# Paths
# subject_id = "SSP011"
# subject_ids = [f'SSP{i:03d}' for i in range(51, 52)] 
directory_path = r"C:\Users\VAH67\Documents\SSP_Cluster\SSP_microgl"

# Check if the directory exists
if os.path.exists(directory_path):
    # Get the list of folder names in the directory
    folder_names = [name for name in os.listdir(directory_path) if os.path.isdir(os.path.join(directory_path, name))]
    print("Folder names:", folder_names)
else:
    print(f"The directory {directory_path} does not exist.")

for subject_id in folder_names:
    # data_path = fr"C:\Users\VAH67\Documents\SSP\data_sample\SSP\{subject_id}_BIDS\microgl_data"
    data_path = fr"C:\Users\VAH67\Documents\SSP_Cluster\SSP_microgl\{subject_id}"
    # bids_base_path = fr"C:\Users\VAH67\Documents\SSP\data_sample\SSP\{subject_id}_BIDS\sub-{subject_id[-3:]}"
    bids_base_path = fr"C:\Users\VAH67\Documents\SSP_Cluster\bids\sub-{subject_id[-3:]}"

    # Ensure the BIDS base directory exists
    os.makedirs(bids_base_path, exist_ok=True)

    # Define the directories for 'anat', 'fmap', and 'func'
    anat_dir = os.path.join(bids_base_path, 'anat')
    fmap_dir = os.path.join(bids_base_path, 'fmap')
    func_dir = os.path.join(bids_base_path, 'func')

    # Create these directories if they don't already exist
    os.makedirs(anat_dir, exist_ok=True)
    os.makedirs(fmap_dir, exist_ok=True)
    os.makedirs(func_dir, exist_ok=True)

    # Regex patterns to identify different types of files
    func_pattern = re.compile(fr"{subject_id}_MR1_(\w+)_fMRI_(AP\d+)_(\d+[a-z]?)\.(nii.gz|json)$")
    rest_pattern = re.compile(fr"{subject_id}_MR1_rfMRI_REST_AP_(\d+[a-z]?)\.(nii.gz|json)$")
    anat_pattern = re.compile(fr"{subject_id}_MR1_ABCD_T1w_MPR_vNav_(\d+)(?:_i(\d+))?\.nii.gz$")
    fmap_phase_pattern = re.compile(fr"{subject_id}_MR1_gre_field_mapping_(\d+)_e(\d+)_ph\.(nii.gz|json)$")
    fmap_magnitude_pattern = re.compile(fr"{subject_id}_MR1_gre_field_mapping_(\d+)_e(\d+)\.(nii.gz|json)$")
    epi_pattern_AP = re.compile(fr"{subject_id}_MR1_SpinEchoFieldMap_AP_(\d+)\.(nii.gz|json)$")
    epi_pattern_PA = re.compile(fr"{subject_id}_MR1_SpinEchoFieldMap_PA_(\d+)\.(nii.gz|json)$")
    sbref_pattern = re.compile(r".*SBRef.*")

    # Counter for anatomical files to ensure unique naming within each sequence
    anat_counter = {}
    max_sequence = 0  # Variable to track the maximum sequence number

    # List to store skipped files
    skipped_files = []

    # Dictionary to store conversion information
    conversion_info = {}

    # IntendedFor list to be added to the JSON files
    intended_for = [
        f"func/sub-{subject_id[-3:]}_task-ALICE_run-1_bold.nii.gz",
        f"func/sub-{subject_id[-3:]}_task-ALICE_run-2_bold.nii.gz",
        f"func/sub-{subject_id[-3:]}_task-BADAGA_run-1_bold.nii.gz",
        f"func/sub-{subject_id[-3:]}_task-BADAGA_run-2_bold.nii.gz",
        f"func/sub-{subject_id[-3:]}_task-BADAGA_run-3_bold.nii.gz",
        f"func/sub-{subject_id[-3:]}_task-rest_run-1_bold.nii.gz"
    ]

    # Process and copy files
    def process_files(directory):
        global max_sequence
        for file in os.listdir(directory):
            source_path = os.path.join(directory, file)
            if file in skipped_files:
                continue

            # Check and skip SBRef files
            if sbref_pattern.match(file):
                skipped_files.append(file)
                continue

            # Handle JSON files for SBRef checking
            if file.endswith('.json'):
                with open(source_path, 'r') as json_file:
                    json_data = json.load(json_file)
                    if 'SeriesDescription' in json_data and 'SBRef' in json_data['SeriesDescription']:
                        nifti_file = file.replace('.json', '.nii.gz')
                        if os.path.exists(os.path.join(directory, nifti_file)):
                            skipped_files.extend([file, nifti_file])
                            continue

            # Handle functional files
            func_match = func_pattern.match(file)
            if func_match:
                task, ap, num, ext = func_match.groups()
                run_number = ap.replace('AP', '')
                bids_filename = f'sub-{subject_id[-3:]}_task-{task}_run-{run_number}_bold.{ext}'
                destination_path = os.path.join(func_dir, bids_filename)
                shutil.copy(source_path, destination_path)
                print(f"Copied {source_path} to {destination_path}")
                conversion_info[file] = bids_filename
                continue

            # Handle Rest files
            rest_match = rest_pattern.match(file)
            if rest_match:
                num, ext = rest_match.groups()
                run_number = 1
                bids_filename = f'sub-{subject_id[-3:]}_task-rest_run-{run_number}_bold.{ext}'
                destination_path = os.path.join(func_dir, bids_filename)
                shutil.copy(source_path, destination_path)
                print(f"Copied {source_path} to {destination_path}")
                conversion_info[file] = bids_filename
                continue

            # Handle anatomical files
            anat_match = anat_pattern.match(file)
            if anat_match:
                sequence, index = anat_match.groups()
                sequence_num = int(sequence)
                if sequence_num > max_sequence:
                    max_sequence = sequence_num  # Update max sequence if current is larger
                
                if sequence not in anat_counter:
                    anat_counter[sequence] = 0
                anat_counter[sequence] += 1
                suffix = anat_counter[sequence]
                index_suffix = f"_{index}" if index else ""
                bids_filename = f'sub-{subject_id[-3:]}_T1w_{sequence}{index_suffix}.nii.gz'
                destination_path = os.path.join(anat_dir, bids_filename)
                shutil.copy(source_path, destination_path)
                print(f"Copied {source_path} to {destination_path}")
                conversion_info[file] = bids_filename
                continue

            # Handle field map files (phase and magnitude)
            fmap_phase_match = fmap_phase_pattern.match(file)
            if fmap_phase_match:
                echo_num, num, ext = fmap_phase_match.groups()
                bids_filename = f'sub-{subject_id[-3:]}_phase{num}.{ext}'
                destination_path = os.path.join(fmap_dir, bids_filename)
                shutil.copy(source_path, destination_path)
                print(f"Copied {source_path} to {destination_path}")
                conversion_info[file] = bids_filename
                continue

            fmap_magnitude_match = fmap_magnitude_pattern.match(file)
            if fmap_magnitude_match:
                echo_num, num, ext = fmap_magnitude_match.groups()
                bids_filename = f'sub-{subject_id[-3:]}_magnitude{num}.{ext}'
                destination_path = os.path.join(fmap_dir, bids_filename)
                shutil.copy(source_path, destination_path)
                print(f"Copied {source_path} to {destination_path}")
                conversion_info[file] = bids_filename
                continue

            # Handle EPI field map files
            epi_match_AP = epi_pattern_AP.match(file)
            if epi_match_AP:
                num, ext = epi_match_AP.groups()
                bids_filename = f'sub-{subject_id[-3:]}_dir-AP_epi.{ext}'
                destination_path = os.path.join(fmap_dir, bids_filename)
                shutil.copy(source_path, destination_path)
                print(f"Copied {source_path} to {destination_path}")
                conversion_info[file] = bids_filename
                # Add IntendedFor to the JSON file
                if ext == 'json':
                    with open(destination_path, 'r+') as json_file:
                        json_data = json.load(json_file)
                        json_data['IntendedFor'] = intended_for
                        json_file.seek(0)
                        json.dump(json_data, json_file, indent=4)
                        json_file.truncate()
                continue

            epi_match_PA = epi_pattern_PA.match(file)
            if epi_match_PA:
                num, ext = epi_match_PA.groups()
                bids_filename = f'sub-{subject_id[-3:]}_dir-PA_epi.{ext}'
                destination_path = os.path.join(fmap_dir, bids_filename)
                shutil.copy(source_path, destination_path)
                print(f"Copied {source_path} to {destination_path}")
                conversion_info[file] = bids_filename
                # Add IntendedFor to the JSON file
                if ext == 'json':
                    with open(destination_path, 'r+') as json_file:
                        json_data = json.load(json_file)
                        json_data['IntendedFor'] = intended_for
                        json_file.seek(0)
                        json.dump(json_data, json_file, indent=4)
                        json_file.truncate()
                continue

    # Call the processing function
    process_files(data_path)

    # Function to get user input for overwrite decision
    def get_overwrite_decision():
        print(f"The file {new_anat_path} already exists. Do you want to overwrite it? (type 'yes' or hit Enter to overwrite, type 'no' or hit Escape to skip): ")
        while True:
            user_input = input().lower()
            if user_input in ['yes', 'y', '']:
                return 'yes'
            elif user_input in ['no', 'n']:
                return 'no'

    # Function to get user input for cleaning up the anat folder
    def get_cleanup_decision():
        print(f"Do you want to keep only the {new_anat_path} file in the anat folder and delete the rest? (type 'yes' or hit Enter to keep only the main file, type 'no' or hit Escape to keep all files): ")
        while True:
            user_input = input().lower()
            if user_input in ['yes', 'y', '']:
                return 'yes'
            elif user_input in ['no', 'n']:
                return 'no'

    # Renaming the highest sequence anatomical file
    highest_anat_file = f'sub-{subject_id[-3:]}_T1w_{max_sequence}.nii.gz'
    highest_anat_path = os.path.join(anat_dir, highest_anat_file)
    new_anat_path = os.path.join(anat_dir, f'sub-{subject_id[-3:]}_T1w.nii.gz')

    # Check if the file already exists
    if os.path.exists(new_anat_path):
        decision = get_overwrite_decision()
        if decision == 'yes':
            os.remove(new_anat_path)
            os.rename(highest_anat_path, new_anat_path)
            conversion_info[highest_anat_file] = f'sub-{subject_id[-3:]}_T1w.nii.gz'
        else:
            print(f"Skipping overwrite of {new_anat_path}.")
    else:
        os.rename(highest_anat_path, new_anat_path)
        conversion_info[highest_anat_file] = f'sub-{subject_id[-3:]}_T1w.nii.gz'

    # Ask if the user wants to keep only the main T1w file
    cleanup_decision = get_cleanup_decision()
    if cleanup_decision == 'yes':
        for file in os.listdir(anat_dir):
            if file != f'sub-{subject_id[-3:]}_T1w.nii.gz':
                os.remove(os.path.join(anat_dir, file))

    # Print skipped files
    print("Skipped files (SBRef):")
    for file in skipped_files:
        print(file)

    # Print converted files
    print("\nConverted files:")
    for original, converted in conversion_info.items():
        print(f"{original} -> {converted}")

    print("\nConversion to BIDS structure complete.")
```

```python
import os
import json
import pandas as pd

# Base paths
bids_cluster_base_path = r"C:\Users\VAH67\Documents\SSP_Cluster\bids"
participants_tsv_path = os.path.join(bids_cluster_base_path, "participants.tsv")

# Initialize a list to store participant data
participants_data = []

# Function to process each subject
def process_subject(subject_id):
    func_dir = os.path.join(bids_cluster_base_path, f"sub-{subject_id}", 'func')
    json_file_path = os.path.join(func_dir, f"sub-{subject_id}_task-ALICE_run-1_bold.json")
    
    if os.path.exists(json_file_path):
        with open(json_file_path, 'r') as json_file:
            json_data = json.load(json_file)
            patient_sex = json_data.get("PatientSex", "N/A")
            patient_age = json_data.get("PatientAge", "0")
            # Convert patient age to integer
            try:
                patient_age = int(float(patient_age))
            except ValueError:
                patient_age = 0  # Default value if conversion fails
            participants_data.append({
                "participant_id": f"sub-{subject_id}",
                "age": patient_age,
                "sex": patient_sex
            })

# Get list of subject folders
subject_folders = [name for name in os.listdir(bids_cluster_base_path) if os.path.isdir(os.path.join(bids_cluster_base_path, name)) and name.startswith('sub-')]

# Process each subject
for subject_folder in subject_folders:
    subject_id = subject_folder.split('-')[1]
    process_subject(subject_id)

# Load existing participants.tsv file
if os.path.exists(participants_tsv_path):
    participants_df = pd.read_csv(participants_tsv_path, sep='\t')
else:
    # Create a new DataFrame if the file doesn't exist
    participants_df = pd.DataFrame(columns=["participant_id", "age", "sex"])

# Create a DataFrame from the new data
new_participants_df = pd.DataFrame(participants_data)

# Merge the existing and new DataFrames
participants_df = pd.concat([participants_df, new_participants_df]).drop_duplicates(subset=["participant_id"]).reset_index(drop=True)

# Fill NaN values in the age column with 0 and convert to integers
participants_df["age"] = participants_df["age"].fillna(0).astype(int)

# Save the updated participants.tsv file
participants_df.to_csv(participants_tsv_path, sep='\t', index=False)

print("participants.tsv file has been updated.")
```
